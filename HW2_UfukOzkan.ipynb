```python

filtered_data = match_data[(match_data['suspended'] == False) & (match_data['stopped'] == False)]
filtered_data['P_home_win'] = 1 / filtered_data['1']
filtered_data['P_draw'] = 1 / filtered_data['X']
filtered_data['P_away_win'] = 1 / filtered_data['2']
filtered_data['P_diff'] = filtered_data['P_home_win'] - filtered_data['P_away_win']
bins = np.linspace(-1, 1, 21)  # Create 20 bins between -1 and 1
filtered_data['P_diff_bin'] = pd.cut(filtered_data['P_diff'], bins)
bin_stats = filtered_data.groupby('P_diff_bin').apply(
    lambda group: pd.Series({
        'Games_in_bin': len(group),
        'Draws_in_bin': (group['result'] == 'X').sum(),
        'Draw_probability': (group['result'] == 'X').mean()
    })
)

# Display the bin statistics


bin_stats.reset_index(inplace=True)
print(bin_stats)
plt.figure(figsize=(12, 8))
plt.bar(bin_stats['P_diff_bin'].astype(str), bin_stats['Draw_probability'], color='blue', alpha=0.7)
plt.title('Draw Probability per P(home win) - P(away win) Bin', fontsize=16)
plt.xlabel('P(home win) - P(away win) Bin', fontsize=14)
plt.ylabel('Draw Probability', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Results

Statistical Summary: The draw probabilities (Draw_probability) increased as the normalized probabilities for home and away wins approached equality (P_diff ≈ 0). This suggests a higher likelihood of draws in matches with balanced odds.

Conclusions:

```python
late_goal_matches = match_data[
    (match_data['minute'] > 90) & 
    (
        ((match_data['current_state'] == 'home_win') & (match_data['result'] == 'draw')) |  # The team ahead draws
        ((match_data['current_state'] == 'away_win') & (match_data['result'] == 'draw')) |  # The team ahead draws
        ((match_data['current_state'] == 'draw') & (match_data['result'] == 'home_win')) |  # The home team wins from a draw
        ((match_data['current_state'] == 'draw') & (match_data['result'] == 'away_win'))    # The away team wins from a draw
    )
]['fixture_id'].unique()

early_red_card_matches = match_data[
    (match_data['minute'] <= 15) & 
    ((match_data['Yellowred Cards - home'] > 0) | (match_data['Yellowred Cards - away'] > 0))
]['fixture_id'].unique()

matches_to_remove = set(late_goal_matches).union(set(early_red_card_matches))
filtered_cleaned_data = match_data[~match_data['fixture_id'].isin(matches_to_remove)]

late_goal_count = len(late_goal_matches)
early_red_card_count = len(early_red_card_matches)
total_removed_matches = len(matches_to_remove)

print(f"90. dakikadan sonra sonucu değiştiren maç sayisi: {late_goal_count}")
print(f"İlk 15 dakikada kirmizi kart görülen maç sayisi: {early_red_card_count}")
print(f"Toplam çikarilan maç sayisi: {total_removed_matches}")




```

```python
# P_diff calculation
filtered_cleaned_data.loc[:,'P_home_win'] = 1 / filtered_cleaned_data['1']
filtered_cleaned_data.loc[:,'P_draw'] = 1 / filtered_cleaned_data['X']
filtered_cleaned_data.loc[:,'P_away_win'] = 1 / filtered_cleaned_data['2']

filtered_cleaned_data.loc[:,'P_diff'] = filtered_cleaned_data['P_home_win'] - filtered_cleaned_data['P_away_win']

bins = np.linspace(-1, 1, 21)  # 20 eşit aralık
filtered_cleaned_data.loc[:,'P_diff_bin'] = pd.cut(filtered_cleaned_data['P_diff'], bins)
```

```python
# Analysis for each bins
bin_stats_cleaned = filtered_cleaned_data.groupby('P_diff_bin').apply(
    lambda group: pd.Series({
        'Games_in_bin': len(group),
        'Draws_in_bin': (group['result'] == 'X').sum(),
        'Draw_probability': (group['result'] == 'X').mean()
    })
)

# Adding bookmaker predictions
filtered_cleaned_data['Implied_draw_prob'] = filtered_cleaned_data['P_draw'] / (
    filtered_cleaned_data['P_home_win'] + filtered_cleaned_data['P_draw'] + filtered_cleaned_data['P_away_win']
)

bin_stats_cleaned['Avg_implied_draw_prob'] = filtered_cleaned_data.groupby('P_diff_bin')['Implied_draw_prob'].mean()
bin_stats_cleaned['Difference'] = bin_stats_cleaned['Draw_probability'] - bin_stats_cleaned['Avg_implied_draw_prob']
```

```python
# "Draw Probability" Visualization
plt.figure(figsize=(12, 8))
plt.bar(bin_stats_cleaned.index.astype(str), bin_stats_cleaned['Draw_probability'], color='blue', alpha=0.7, label="Actual Draw Probability")
plt.bar(bin_stats_cleaned.index.astype(str), bin_stats_cleaned['Avg_implied_draw_prob'], color='orange', alpha=0.7, label="Implied Draw Probability")
plt.title('Draw Probability per P(home win) - P(away win) Bin (Cleaned Data)', fontsize=16)
plt.xlabel('P(home win) - P(away win) Bin', fontsize=14)
plt.ylabel('Draw Probability', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

comparison['Home Win Diff'] = comparison['Predicted Home Win Prob'] - comparison['Implied Home Win Prob']
comparison['Draw Diff'] = comparison['Predicted Draw Prob'] - comparison['Implied Draw Prob']
comparison['Away Win Diff'] = comparison['Predicted Away Win Prob'] - comparison['Implied Away Win Prob']

largest_positive_diff = comparison.nlargest(10, 'Home Win Diff')
largest_negative_diff = comparison.nsmallest(10, 'Home Win Diff')
largest_positive_draw = comparison.nlargest(10, 'Draw Diff')
largest_negative_draw = comparison.nsmallest(10, 'Draw Diff')

largest_positive_diff.to_csv('largest_positive_diff_home.csv', index=False)
largest_negative_diff.to_csv('largest_negative_diff_home.csv', index=False)
largest_positive_draw.to_csv('largest_positive_diff_draw.csv', index=False)
largest_negative_draw.to_csv('largest_negative_diff_draw.csv', index=False)

plt.figure(figsize=(14, 6))
sns.histplot(comparison['Home Win Diff'], kde=True, bins=30, color='blue', label='Home Win Diff')
sns.histplot(comparison['Draw Diff'], kde=True, bins=30, color='orange', label='Draw Diff')
sns.histplot(comparison['Away Win Diff'], kde=True, bins=30, color='green', label='Away Win Diff')
plt.axvline(0, color='red', linestyle='--', label='No Difference')
plt.title('Distribution of Probability Differences')
plt.xlabel('Difference (Predicted - Implied)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
comparison['Prediction Match'] = (
    (comparison['Predicted Home Win Prob'] > 0.5) & (comparison['Implied Home Win Prob'] > 0.5) |
    (comparison['Predicted Draw Prob'] > 0.5) & (comparison['Implied Draw Prob'] > 0.5) |
    (comparison['Predicted Away Win Prob'] > 0.5) & (comparison['Implied Away Win Prob'] > 0.5)
)
comparison['P_diff_bin'] = pd.cut(comparison['P_diff'], bins=10)
mismatch_analysis = comparison.groupby('P_diff_bin')['Prediction Match'].mean()

plt.figure(figsize=(14, 6))
mismatch_analysis.plot(kind='bar', color='purple')
plt.title('Mismatch Rates by P_diff Bins')
plt.xlabel('P_diff Bin')
plt.ylabel('Proportion of Correct Predictions')
plt.xticks(rotation=45)
plt.show()

significant_differences = comparison[
    (comparison['Home Win Diff'].abs() > 0.2) |
    (comparison['Draw Diff'].abs() > 0.2) |
    (comparison['Away Win Diff'].abs() > 0.2)
]

significant_differences.to_csv('significant_differences.csv', index=False)

# Explanation:
# - The `nlargest` and `nsmallest` functions identify matches with the highest and lowest differences.
# - `sns.histplot` visualizes distributions of probability differences for Home Win, Draw, and Away Win.
# - Mismatch analysis groups 'P_diff' into bins and computes the mismatch rate for each bin.
# - Rows with significant probability differences (e.g., > 0.2) are saved for further investigation.
```
